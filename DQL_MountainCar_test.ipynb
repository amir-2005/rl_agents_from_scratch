```python
import gymnasium as gym
import matplotlib.pyplot as plt
from DeepQLearning import DQL
```

```python
env_car = gym.make("MountainCar-v0")
```

```python
agent = DQL(
    env_car,
    hidden_layers=(64, 64),
    activation="Tanh",
    learning_rate=1e-3,
    epsilon_decay=0.995,
    discount_factor=0.98,
    batch_size=64,
    buffer_size=20000
)

rewards = agent.train(target_update=500, n_episodes=4000)
```

```python
plt.bar(x=range(len(rewards)), height=rewards)
plt.ylabel("Reward")
plt.xlabel("Episode");
```

```python
env_car_test = gym.make("MountainCar-v0", render_mode="human")
agent.test(env_car_test, 3)
```
